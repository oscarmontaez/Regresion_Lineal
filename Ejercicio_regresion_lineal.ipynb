```python
#Ejercicio de Regresion Lineal
#Importamos las librerias
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets # usando sklear para saber los valores optimos
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Carga del archivo con la base de datos
data = pd.read_csv("C:/Users/oscar/Documents/DataSets/data.csv", sep=",",index_col=False, header=0)
x = data.metro.values
y = data.precio.values
```

```python
# Exploraci√≥n de los datos
data.columns
data.info
data.describe
```

```python
#Grafica de la data
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
# Aplicacion de la Regresion
regresion = linear_model.LinearRegression()
```

```python
metros = data["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metros, data["precio"])
```

```python
print("Interseccion (b)", modelo.intercept_)
#imprimos la pendiente
print("Pendiente (m)", modelo.coef_)
```

```python
entrada= [[20],[35]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label='Datos originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='L√≠nea de regresi√≥n')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
